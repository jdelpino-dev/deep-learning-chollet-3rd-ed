```
input_dim = 2
output_dim = 1
W = tf.Variable(initial_value=tf.random.uniform(shape=(input_dim, output_dim)))
b = tf.Variable(initial_value=tf.zeros(shape=(output_dim,)))
```

```
def model(inputs, W, b):
    return tf.matmul(inputs, W) + b
```

```
def mean_squared_error(targets, predictions):
    per_sample_losses = tf.square(targets - predictions)
    return tf.reduce_mean(per_sample_losses)
```

```
learning_rate = 0.1

@tf.function(jit_compile=True)
def training_step(inputs, targets, W, b):
    with tf.GradientTape() as tape:
        predictions = model(inputs, W, b)
        loss = mean_squared_error(predictions, targets)
    grad_loss_wrt_W, grad_loss_wrt_b = tape.gradient(loss, [W, b])
    W.assign_sub(grad_loss_wrt_W * learning_rate)
    b.assign_sub(grad_loss_wrt_b * learning_rate)
    return loss
```

```
for step in range(40):
    loss = training_step(inputs, targets, W, b)
    print(f"Loss at step {step}: {loss:.4f}")
```

```
input_dim = 2
output_dim = 1

W = torch.rand(input_dim, output_dim, requires_grad=True)
b = torch.zeros(output_dim, requires_grad=True)
```

```
def model(inputs, W, b):
    return torch.matmul(inputs, W) + b
```

```
def mean_squared_error(targets, predictions):
    per_sample_losses = torch.square(targets - predictions)
    return torch.mean(per_sample_losses)
```

```
learning_rate = 0.1

def training_step(inputs, targets, W, b):
    predictions = model(inputs)
    loss = mean_squared_error(targets, predictions)
    loss.backward()
    grad_loss_wrt_W, grad_loss_wrt_b = W.grad, b.grad
    with torch.no_grad():
        W -= grad_loss_wrt_W * learning_rate
        b -= grad_loss_wrt_b * learning_rate
    W.grad = None
    b.grad = None
    return loss
```

#### What makes the JAX approach unique

#### A Clean, Non-Global JAX Training Script

In the previous examples, it’s easy to accidentally rely on global variables—learning_rate, grad_fn, and even dimension constants—because short tutorial snippets optimize for brevity over structure. That’s fine for a page in a book, but it becomes fragile in real work: functions hide dependencies, small edits break reproducibility, and jit can quietly capture Python values as static constants.

This section rewrites the whole JAX training flow in a clean, functional JAX style: all configuration and model state are explicit, parameters are passed as a (W, b) tuple, and each step returns a new state instead of mutating globals. The training step is built via a small factory (make_training_step) so the learning rate is explicitly bound where it belongs, while keeping the JIT-compiled core pure and reusable. The result is a notebook-friendly script that’s easier to debug, easier to test, and safer to extend (e.g., swapping optimizers, adding regularization, batching, or moving to a real model library later).

```
# Cell 1: Imports
# - NumPy: for simple RNG / data handling
# - JAX: autodiff + JIT compilation
# - jax.numpy: NumPy-like API backed by JAX

import numpy as np
import jax
import jax.numpy as jnp
```

```
# Cell 2: Forward-pass building blocks
# - dense(): example "layer" with ReLU (not used by the linear model below, but kept for reference)
# - model(): linear model y = XW + b

@jax.jit
def dense(inputs, W, b):
    # JIT compiles this function for faster repeated execution
    return jax.nn.relu(jnp.matmul(inputs, W) + b)  # ReLU(XW + b)

def model(inputs, W, b):
    # Linear model: predictions = XW + b
    return jnp.matmul(inputs, W) + b
```

```
# Cell 3: Loss function
# Mean Squared Error (MSE) is scalar-valued and differentiable.

def mean_squared_error(targets, predictions):
    per_sample_losses = jnp.square(targets - predictions)  # (y - y_hat)^2 for each sample
    return jnp.mean(per_sample_losses)                      # average across samples
```

```
# Cell 4: Compute loss from a parameter "state"
# We bundle parameters into a tuple (W, b) to keep things functional (no globals, no mutation).

def compute_loss(state, inputs, targets):
    W, b = state                                # unpack parameters
    predictions = model(inputs, W, b)           # forward pass
    loss = mean_squared_error(targets, predictions)  # compute scalar loss
    return loss
```

```
# Cell 5: Build a training_step function with a chosen learning rate
# - We create grad_fn inside to avoid global variables
# - training_step is JIT-compiled and returns (loss, new_state)

def make_training_step(learning_rate):
    grad_fn = jax.value_and_grad(compute_loss)  # returns (loss, grads) where grads matches state structure

    @jax.jit
    def training_step(state, inputs, targets):
        loss, grads = grad_fn(state, inputs, targets)  # grads is (dW, db)
        dW, db = grads                                 # unpack gradients
        W, b = state                                   # unpack parameters

        # Gradient descent update (pure/functional: create new arrays, no in-place mutation)
        new_state = (W - learning_rate * dW, b - learning_rate * db)

        return loss, new_state

    return training_step
```

```
# Cell 6: Training loop
# - Initializes W (and optionally any other stochastic components) using JAX PRNG keys.
#   Why JAX keys instead of NumPy RNG?
#   * JAX randomness is explicit and functional: random draws are pure functions of a key.
#   * This composes cleanly with @jit / vmap / pmap and avoids hidden global RNG state.
#   * It’s more future-proof: if you later add dropout/data augmentation/noise, you can
#     split and thread keys through training deterministically.
#   * It avoids generating random values on the host with NumPy and then copying to device.
# - Keeps parameters as JAX arrays (DeviceArrays) from the start.
# - Repeatedly calls the JIT-compiled training_step and prints the loss.

def train(inputs, targets, input_dim, output_dim, learning_rate, steps, seed=0):
    # Create a JAX PRNG key from an integer seed (JAX-style reproducibility)
    key = jax.random.key(seed)

    # Split the key so each random draw uses a distinct, deterministic subkey
    key_W, key_b = jax.random.split(key, 2)

    # Initialize parameters directly in JAX (on-device friendly)
    W0 = jax.random.uniform(key_W, shape=(input_dim, output_dim), dtype=jnp.float32)
    b0 = jnp.zeros((output_dim,), dtype=jnp.float32)  # deterministic; no key needed
    state = (W0, b0)

    training_step = make_training_step(learning_rate)

    for step in range(steps):
        loss, state = training_step(state, inputs, targets)
        print(f"Loss at step {step}: {loss:.4f}")

    return state
```

```
# Cell 7: Prepare inputs/targets for JAX
# Make sure inputs and targets are JAX arrays (DeviceArrays) with float32 dtype.
# Assumes you already have `inputs` and `targets` as NumPy arrays.
#
# Performance win:
# Converting once up front avoids repeatedly passing NumPy (host) arrays into a @jit-compiled
# training loop, which can trigger repeated host→device transfers / staging overhead each step
# (especially noticeable on GPU/TPU). Keeping inputs/targets as JAX arrays lets JAX reuse the
# same device buffers across iterations and keeps shapes/dtypes stable for JIT caching.

inputs_j = jnp.array(inputs, dtype=jnp.float32)   # shape: (N, 2)
targets_j = jnp.array(targets, dtype=jnp.float32) # shape: (N, 1)

# (Optional) sanity checks
print("inputs_j:", inputs_j.shape, inputs_j.dtype)
print("targets_j:", targets_j.shape, targets_j.dtype)
```

```
# Cell 8: Run training
# This calls train() and returns the final learned parameters (W, b).

final_state = train(
    inputs_j,
    targets_j,
    input_dim=2,
    output_dim=1,
    learning_rate=0.1,
    steps=86,
    seed=0,
)

W_final, b_final = final_state
print("W_final:\n", W_final)
print("b_final:\n", b_final)
```

```
predictions = model(inputs_j, W_final, b_final)
plt.figure(figsize=(8, 6))
plt.scatter(inputs_j[:, 0], inputs_j[:, 1], c=predictions[:, 0] > 0.5, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('JAX Linear Classifier Predictions')
plt.colorbar(label='Predicted Class')
plt.show()

# Plot decision boundary
x = jnp.linspace(-1, 4, 100)
y = -W_final[0] / W_final[1] * x + (0.5 - b_final) / W_final[1]
plt.figure(figsize=(8, 6))
plt.plot(x, y, "-r", linewidth=2, label='Decision Boundary')
plt.scatter(inputs_j[:, 0], inputs_j[:, 1], c=predictions[:, 0] > 0.5, cmap='viridis', alpha=0.6)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('JAX Linear Classifier with Decision Boundary')
plt.legend()
plt.show()
```

#### Layers: The building blocks of deep learning

##### The base `Layer` class in Keras

```
import keras

class SimpleDense(keras.Layer):
    def __init__(self, units, activation=None):
        super().__init__()
        self.units = units
        self.activation = activation

    def build(self, input_shape):
        batch_dim, input_dim = input_shape
        self.W = self.add_weight(
            shape=(input_dim, self.units), initializer="random_normal"
        )
        self.b = self.add_weight(shape=(self.units,), initializer="zeros")

    def call(self, inputs):
        y = keras.ops.matmul(inputs, self.W) + self.b
        if self.activation is not None:
            y = self.activation(y)
        return y
```
